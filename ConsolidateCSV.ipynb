```python
#bring in the dependencies
import pandas as pd
import glob
```

```python
files = glob.glob('basic_csv/*')
len(files)
```

```python
file_list = []

for filename in files:
    df = pd.read_csv(filename, index_col=None, header=0)
    file_list.append(df)

tweet_df = pd.concat(file_list, axis=0, ignore_index=False, sort=False)
```

```python
tweet_df.head()
```

```python
tweet_df.to_csv(r'master_csv\tweet_df.csv')
```
