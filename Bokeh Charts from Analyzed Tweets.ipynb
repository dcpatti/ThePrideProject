```python
#start with all of the imports
import pandas as pd
import glob
import json
import numpy as np
import pandas as pd
import csv
import os
import re
import seaborn as sns
import matplotlib.pyplot as plt
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral5
from bokeh.plotting import figure
from bokeh.sampledata.autompg import autompg as df
from bokeh.transform import factor_cmap
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import FuncTickFormatter
from bokeh.models import HoverTool
```

```python
#get the data from the spreadsheet and set the index on the Team column
df=pd.read_csv("reshaped_grouped_2.csv",index_col=0)
df.head()
```

```python
chart_df =df[['Most_Negative','Moderate_Negative','Neutral','Moderate_Positive','Most_Positive','Anchor']]
chart_df.head()
```

```python
#next we need to make a new DF so we can add an index and move the existing index to a column
df2=chart_df.reset_index()
df2.head()
```

```python
output_notebook()

# Store the data in a ColumnDataSource
tweets_cds = ColumnDataSource(df2)

# Specify the selection tools to be made available
select_tools = ['box_select', 'lasso_select', 'poly_select', 'tap', 'reset']

# Create the figure
fig = figure(plot_height=400,
             plot_width=600,
             x_axis_label='Team',
             y_axis_label='Count of Tweets',
             title='Mapping the Highly Polarized Tweets',
             toolbar_location='below',
             tools=select_tools)

# Format the y-axis tick labels as percentages
#fig.yaxis[0].formatter = NumeralTickFormatter(format='00.0%')

# Add marker for the most positive tweets
fig.circle(x='Anchor',
           y='Most_Positive',
           source=tweets_cds,
           size=15,
           color='green',
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)

#add marker for moderately positive tweets
fig.circle(x='Anchor',
           y='Moderate_Positive',
           source=tweets_cds,
           size=15,
           color='lightgreen',
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)

#add marker for neutral tweets
fig.circle(x='Anchor',
           y='Neutral',
           source=tweets_cds,
           size=15,
           color='yellow',
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)

#add marker for moderately negative tweets
fig.circle(x='Anchor',
           y='Moderate_Negative',
           source=tweets_cds,
           size=15,
           color='orange',
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)


# Add marker for the most negative tweets
fig.circle(x='Anchor',
           y='Most_Negative',
           source=tweets_cds,
           color='red',
           size=15,
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)

# Format the tooltip
tooltips = [
            ('Team: ','@teams'),
            ('# Strongly Negative', '@Most_Negative'),
            ('# Strongly Positive', '@Most_Positive')
            
           ]

# Add the HoverTool to the figure
fig.add_tools(HoverTool(tooltips=tooltips))

# Visualize
show(fig)
```
